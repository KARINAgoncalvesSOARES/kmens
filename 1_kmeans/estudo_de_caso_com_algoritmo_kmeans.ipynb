# 1. Atividade prátiva
### Análise das características dos clusters gerados e relacionamento com as regras geradas pelo apriori.

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Carregamento dos dados em dataframe
df = pd.read_csv('/content/drive/MyDrive/A4.csv')
```

```python
df.head(15)
```

```python
df.shape
```

```python
# Execução do algoritmo k-means para diferentes valores de k

rms = []
for k in range(1,23):
  kmeans = KMeans(n_clusters=k, max_iter=1000)
  kmeans.fit(df)
  rms.append(kmeans.inertia_)

# Traçando gráfico do joelho
plt.plot(range(1, 23), rms, marker='o')
plt.title('Gráfico Do Joelho')
plt.xlabel('Número de Clusters')
plt.ylabel('Decaimento do erro RMS')
plt.show()
```

### Utilizando a base de dados "A4.csv". Foi executado o experimento com o kmeans no Weka e K-Means Python usando IDE Google colab e verificou-se que a melhor quantidade de clusters gerados pelo modelo são sete(7). Além do gráfico mostrado acima. È possivel verificar que a curva do joelho ocorre no terceiro cluster.

```python
# Carregamento dos dados em dataframe
df = pd.read_csv('/content/drive/MyDrive/iris.csv')
```

```python
# Mostrando o Dataframe original
df.head(3)
```

```python
df.shape
```

```python
# A coluna variety contém valores categóricos
df['variety'].unique()
```

```python
# Criando função para converter os dados categoricos da coluna variety para dados numéricos

def strg_para_num(df):

  # Definir o mapa de conversão de valores categóricos para numéricos
  mapa = {'Setosa':0, 'Versicolor':1, 'Virginica':2}

  # Fazer a substituição dos valores categóricos pelos valores numéricos correspondentes
  df['variety_num'] = df['variety'].replace(mapa)

  # Retornar o DataFrame com a nova coluna de valores numéricos
  return df
```

```python
# Novo dataframe com dados categóricos convertidos para dados numéricos

df1 = strg_para_num(df)
df1
```

```python
# Aqui observamos que os dados da coluna variety_num são dados numéricos
df['variety_num'].unique()
```

```python
# Excluindo coluna com os dados categóricos
df.drop(['variety'], axis=1, inplace=True)
```

```python
df.head()
```

```python
rms = []
for k in range(1,23):
  kmeans = KMeans(n_clusters=k, max_iter=1000)
  kmeans.fit(df)
  rms.append(kmeans.inertia_)

# Traçando gráfico do joelho
plt.plot(range(1, 23), rms, marker='o')
plt.title('Gráfico Do Joelho')
plt.xlabel('Número de Clusters')
plt.ylabel('Decaimento do erro RMS')
plt.show()
```

## No gráfico acima é feita a representação das espécies da flor iris e a visualização do decaimento do erro RMS, onde podemos notar que o ponto de significância é reprensentada no número 3 onde temos 3 clusters para três espécies.
